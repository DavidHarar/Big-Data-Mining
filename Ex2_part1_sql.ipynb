WRITE YOUR SOLUTION BELOW:

For the `the_bachelor` table:

```python
# 1. Add column names to first table 
query_add_firstN = """ALTER TABLE the_bachelor ADD first_name text;"""
conn.execute(query_add_firstN)
conn.commit()
query_add_lastI = """ALTER TABLE the_bachelor ADD last_initial text;"""
conn.execute(query_add_lastI)
conn.commit()
```

```python
# 2. Populate new columns for first table 
query_update_firstN = """ 
  UPDATE the_bachelor 
    SET first_name = CASE 
                      WHEN Name LIKE '% %' THEN UPPER(substr(Name, 1, instr(Name, ' ') - 1))
                      ELSE UPPER(Name)
                      END 
                      """

conn.execute(query_update_firstN)
conn.commit()

query_update_lastI = """ 
  UPDATE the_bachelor 
    SET last_initial = CASE 
                    WHEN Name LIKE '% %' THEN UPPER(SUBSTR(SUBSTR(Name, instr(Name, ' ') + 1), 1, 1))
                    ELSE ''
                    END 
  """
conn.execute(query_update_lastI)
conn.commit()

# Verifying:
cursor = conn.execute('''SELECT * from the_bachelor limit 10''')
for row in cursor:
  print(row)
```
